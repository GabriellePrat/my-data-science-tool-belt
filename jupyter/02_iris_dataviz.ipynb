# Iris Dataset

```python
# Data Analysis
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Disable warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Tune the visual settings for figures in `seaborn`
sns.set_context(
    "notebook", 
    font_scale=1.5,       
    rc={ 
        "figure.figsize": (11, 8), 
        "axes.titlesize": 18 
    }
)

from matplotlib import rcParams
rcParams['figure.figsize'] = 11, 8
```

```python
data = sns.load_dataset('iris')
```

```python
data.head()
```

###Â data visualizations

```python
data.plot(kind="scatter", x="sepal_length", y="sepal_width")
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=data, size=5)
```

```python
sns.FacetGrid(data, hue="species", size=5) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()
```

```python
sns.boxplot(x="species", y="petal_width", data=data)
```

```python
ax = sns.boxplot(x="species", y="petal_width", data=data)
ax = sns.stripplot(x="species", y="petal_width", data=data, jitter=True, edgecolor="gray")
```

```python
sns.violinplot(x="species", y="petal_length", data=data, size=6)
```

```python
sns.FacetGrid(data, hue="species", size=6) \
   .map(sns.kdeplot, "petal_length") \
   .add_legend()
```

```python
sns.pairplot(data, hue="species", size=3)
```

```python
sns.pairplot(data, hue="species", size=3, diag_kind="kde")
```

```python
data.boxplot(by="species", figsize=(12, 6))
```

```python
from pandas.plotting import andrews_curves
andrews_curves(data, "species")
```

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(data, "species")
```

```python
from pandas.plotting import radviz
radviz(data, "species")
```
